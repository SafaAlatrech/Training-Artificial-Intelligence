```python
#___________Question 2_____________#

n = int(input("Enter an integer: "))                 #  getting user input

def fact(a):
	s = 1                                            #  declaring s with the value 1 to use it for Multiplication
	for i in range(1,a+1):                           #  looping n times
		s = s * i                                    #  Multiplying s by i
	return s                                         #  returning the fact as a variale
 
print(fact(n))                                       #  printing the factorial of n

```

```python
#___________Question 4_____________#
word = str(input('Etner a word: '))                   #  getting user input_1 string

n = int(input("Enter a number: "))                    #  getting user input_2 for a integer



while n not in range(0,len(word)):                    #  repeating the onput_2 untill n is in the range of 0, len(word)-1
	n = int(input("Enter a number: "))

def remove(w,a):                                      #  defining a function that will replace a certain	index of a string with an empty cahr ""
	new_word = w.replace(w[a],'',1)                   #  making new variable to hold the new word
	return new_word                                   #  returning the new word


print(remove(word,n))                                 # printing the new word
```

```python
#___________Question 6_____________#
import numpy as np                                   #  importing the numpy library as np

my_array = []                                        #  declaring an empty list_1

a = int(input("Size of array:"))                     #  varaible a holding the size of the first array
for i in range(a):                                   #  looping a times
    my_array.append(int(input("Element:")))          #  adding the elemnts to the first list
my_array = np.array(my_array)                        # converting it into a numpy array

my_array2 = []                                       #  declaring an empty list_2

b = int(input("Size of array:"))                     #  varaible a holding the size of the second array
for i in range(b):                                   #  looping b times
    my_array2.append(int(input("Element:")))         #  adding the elemnts to the second list
my_array2 = np.array(my_array2)                      # converting it into a numpy array

print("\n the covariance between these two arrays:\n",np.cov(my_array, my_array2))                   #  printing the covariance between the first and second numpy arrays
```

```python
#___________Question 7_____________#
import math                                         #  importing the math library
val = input("Enter comma seprated integers : ")     #  putting number in a varaible
D = val.split(",")                                  #  creating a list containing the the values of the variable val with each value beeing superated from the other wiht a comma ","

l = []
C = 50
H = 30

for i in range(0,len(D)):                          #  looping x times  x = length of the array
	l.append(int(math.sqrt((2*C*int(D[i]))/H)))    #  applaying the equation given
	

for i in range(0,len(l)):                          #  printing the values
 	
 	if i == len(l)-1:
 		print(l[i])
 	else:
 		print(l[i],",", end='')
```
